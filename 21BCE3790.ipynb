```python
df.dtypes
```

```python
df['price'] = df['price'].astype(str)
print(df['price'].str.isnumeric().value_counts())
```

```python
df['price'].loc[df['price'].str.isnumeric() == False]
```

```python
df["price"] = pd.to_numeric(df["price"])
mean = df["price"].mean()
median =df["price"].median()
mode = df["price"].mode()
print(mean , median, mode)
```

```python
df.price.value_counts().nlargest(30).plot(kind='bar', figsize=(10,6))
plt.title("Number of boats by price")
plt.ylabel('Number of Boats')
plt.xlabel('Price of the boats')
```

```python
sns.boxplot(x="price",data=df)
plt.show()
```

```python
plt.scatter(df["price"], df["pboat"])
plt.title("Scatter Plot for pboat vs price")
plt.xlabel("pboat")
plt.ylabel("price")
```

```python
sns.boxplot(x="mode",y="price",data=df)
plt.show()
```

```python
sns.pairplot(df,vars = ['catch', 'price','pboat'], kind="reg")
plt.show()
```

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(df,height=2,vars = ['catch', 'price','pboat'], hue="mode")
plt.show()
```

```python
from scipy import stats

corr = stats.pearsonr(df["price"], df["pboat"])
print("p-value:\t", corr[1])
print("cor:\t\t", corr[0])
```

```python
correlation = df.corr(method='pearson', numeric_only=True)
correlation
```

```python
sns.heatmap(correlation,xticklabels=correlation.columns,
            yticklabels=correlation.columns)
```
